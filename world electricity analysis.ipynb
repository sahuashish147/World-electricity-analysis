```python
last = list(df.columns[11:])
col = [df.columns[0]]
col.extend(last)
df = df[col]
x = df[df.columns[1:]]
new_list = []
for i in range(df.shape[0]):
    i *= 100000
    new_list.append(i)
x['new_column'] = new_list
from sklearn.impute import KNNImputer
knn = KNNImputer(n_neighbors = 2)
y = knn.fit_transform(x)
df[df.columns[1:]]  = y[:,:-1]
```
